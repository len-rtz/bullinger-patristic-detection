```python
from locisimiles import ClassificationPipelineWithCandidategeneration
from locisimiles import Document

# Load documents
query = Document("VD_bdc_chunks.csv")
source = Document("VD_psc_chunks.csv")

# Define pipeline
pipeline = ClassificationPipelineWithCandidategeneration(
    classification_name="julian-schelb/philberta-class-lat-intertext-v2",
    embedding_model_name="julian-schelb/multilingual-e5-large-emb-lat-intertext-v1",
    device="cuda",  # or "cuda", "mps"
)

# Run pipeline
results = pipeline.run(query=query, source=source, top_k=100)
```

```python
# use standalone utility functions
from locisimiles.pipeline import results_to_csv, results_to_json

results_to_csv(results, "results-locisimiles-e5-phil-v2.csv")
results_to_json(results, "results-locisimiles-e5-phil-v2.json")
```
