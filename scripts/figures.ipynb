```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from io import StringIO
```

```python
CSV_PATH = "../evaluation/evaluation_results_laberta_vulg_top1000.csv"

df = pd.read_csv(CSV_PATH)
pct_cols = ["Overall_Recall","Explicit_Recall","Implicit_Recall","Overall_MRR","Explicit_MRR","Implicit_MRR"]
df[pct_cols] = df[pct_cols] * 100

sns.set_theme(style="whitegrid", font="serif")
PALETTE = {"Overall": "#0B0B0B", "Explicit": "#C11A9A", "Implicit": "#F27209"}
MARKERS = {"Overall": "o", "Explicit": "s", "Implicit": "^"}
LINE_KW = dict(linewidth=2.2, markersize=8, markeredgewidth=1.4, markeredgecolor="white")

fig, ax = plt.subplots(1, 1, figsize=(7, 5))

# Recall@k
for label, col in [("Overall","Overall_Recall"),("Explicit","Explicit_Recall"),("Implicit","Implicit_Recall")]:
    ax.plot(df["k"], df[col], label=label, color=PALETTE[label], marker=MARKERS[label], **LINE_KW)
    ax.annotate(f"{df[col].iloc[-1]:.0f}%", xy=(df["k"].iloc[-1], df[col].iloc[-1]),
                xytext=(12,0), textcoords="offset points", va="center",
                fontsize=9, color=PALETTE[label], fontweight="bold")

ax.axhline(y=40, color="#2C5F8A", linestyle=":", linewidth=1.4, alpha=0.55)
ax.set_title("Recall@k", fontsize=12, fontweight="semibold", pad=10)
ax.set_xlabel("k (log scale)", fontsize=10)
ax.set_ylabel("Recall (%)", fontsize=10)
ax.set_xscale("log")
ax.set_xticks(df["k"])
ax.get_xaxis().set_major_formatter(mticker.ScalarFormatter())
ax.set_ylim(0, 100)
ax.yaxis.set_major_formatter(mticker.PercentFormatter(decimals=0))
ax.grid(axis="y", alpha=0.35, linestyle="--")
ax.grid(axis="x", alpha=0.2, linestyle=":")


# Create one shared legend under the suptitle
handles, labels = ax.get_legend_handles_labels()

fig.legend(
    handles, labels,
    loc="upper center",
    bbox_to_anchor=(0.5, 0.93),  
    ncol=3,
    frameon=False,
    fontsize=10
)

plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.savefig("../figures/recall_at_k_laberta.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch

models = {
    "LaBERTa\n(bi-encoder)": {
        "path": "../retrieval/VD/laberta-vulg/retrieval_results_laberta_vulg.pkl",
        "type": "biencoder",
        "color": "#3D3D3D",
    },
    "SPhilBERTa-intertext\n(bi-encoder)": {
        "path": "../retrieval/VD/sphilberta-intertext/retrieval_results_sphilberta-intertext.pkl",
        "type": "biencoder",
        "color": "#3D3D3D",
    },
     "e5-intertext\n(bi-encoder)": {
        "path": "../retrieval/VD/e5-intertext/retrieval_results_e5-intertext.pkl",
        "type": "biencoder",
        "color": "#3D3D3D",
    },
        "LaBERTa + PhilBERTa-intertext\n(bi-enc + cross-enc)": {
        "path": "../retrieval/VD/laberta-vulg-philberta/retrieval_results_laberta_vulg_philberta_reranked.pkl",
        "type": "probability",         
        "color": "#3D3D3D",
    },
    "LociSimiles SPhilberta-intertext + Philberta-intertext \n(bi-enc + cross-enc)": {
        "path": "../retrieval/VD/locisimiles-sphil-phil-v1/retrieval_results_locisimiles_judgement_k100.pkl",
        "type": "probability",  
        "color": "#3D3D3D",
    },
    "LociSimiles e5-intertext + Philberta-intertext\n(bi-enc + cross-enc)": {
        "path": "../retrieval/VD/locisimiles-e5-phil-v2/retrieval_results_locisimiles_judgement_e5-phil-v2.pkl",
        "type": "probability",  
        "color": "#3D3D3D",
    },
}

# Load scores to probabilities or cosine 
for name, cfg in models.items():
    with open(cfg["path"], "rb") as f:
        data = pickle.load(f)

    scores = [s for cands in data.values() for _, s in cands]

    if cfg["type"] == "probability":
        cfg["score_label"] = "Classifier probability P(positive)"
    else:
        cfg["score_label"] = "Cosine similarity"

    cfg["scores"] = scores
    cfg["data"]   = data

sns.set_theme(style="whitegrid", font="serif")
fig, axes = plt.subplots(2, 3, figsize=(20, 8))
axes = axes.flatten()

biencoder_models    = {n: c for n, c in models.items() if c["type"] == "biencoder"}
probability_models  = {n: c for n, c in models.items() if c["type"] == "probability"}

bienc_xmin  = min(min(c["scores"]) for c in biencoder_models.values())
bienc_xmax  = max(max(c["scores"]) for c in biencoder_models.values())
bienc_ymax  = max(
    np.histogram(c["scores"], bins=100,
                 range=(bienc_xmin, bienc_xmax))[0].max()
    for c in biencoder_models.values()
)

prob_xmin, prob_xmax = 0.0, 1.0
prob_ymax = max(
    np.histogram(c["scores"], bins=100,
                 range=(prob_xmin, prob_xmax))[0].max()
    for c in probability_models.values()
)

for i, (name, cfg) in enumerate(models.items()):
    scores = cfg["scores"]
    ax     = axes[i]

    if cfg["type"] == "biencoder":
        xmin, xmax, ymax = bienc_xmin, bienc_xmax, bienc_ymax
    else:
        xmin, xmax, ymax = prob_xmin, prob_xmax, prob_ymax

    ax.hist(scores, bins=100, range=(xmin, xmax),
            color=cfg["color"], alpha=0.8, edgecolor="white", linewidth=0.3)
    ax.axvline(np.median(scores), color="black", linestyle="--",
               linewidth=1.2, alpha=0.6, label=f"Median={np.median(scores):.2f}")

    ax.set_xlim(xmin, xmax)
    ax.set_ylim(0, ymax * 1.05)          # 5% headroom
    ax.set_xlabel(cfg["score_label"], fontsize=9)
    ax.set_ylabel("Count", fontsize=9)
    ax.set_title(f"{name} — score distribution", fontsize=10, fontweight="semibold")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig("../figures/score_distributions_Laberta_vs_Loci-2.png", bbox_inches="tight", dpi=300)
plt.show()

print(f"{'Model':<45} {'Min':>7} {'Median':>7} {'Max':>7}  Type")
print("-" * 75)
for name, cfg in models.items():
    s = cfg["scores"]
    print(f"{name.replace(chr(10), ' '):<45} {min(s):>7.3f} {np.median(s):>7.3f} {max(s):>7.3f}  {cfg['type']}")
```

```python
import json
from collections import defaultdict

FATHER_NAMES = {
    'P18700': 'Augustinus',
    'P17746': 'Tertullianus',
    'P18993': 'Ambrosius',
    'P18986': 'Hieronymus',
    'P18048': 'Hilarius',
    'P18988': 'Cyprian'
}

TARGET_FATHERS = set(FATHER_NAMES.keys())

with open("../data/ir_ground_truth.json") as f:
    data = json.load(f)

counts = defaultdict(lambda: {"explicit": 0, "implicit": 0})

for ann in data["ground_truth"]:
    father_id = ann.get("church_fathers")
    ref_type = ann.get("reference_type")
    if father_id in TARGET_FATHERS and ref_type in ("explicit", "implicit"):
        counts[FATHER_NAMES[father_id]][ref_type] += 1

print(f"{'Church Father':<16} {'Explicit':>8} {'Implicit':>8} {'Total':>6}")
print("-" * 40)
for name, c in sorted(counts.items()):
    total = c['explicit'] + c['implicit']
    print(f"{name:<16} {c['explicit']:>8} {c['implicit']:>8} {total:>6}")
```
