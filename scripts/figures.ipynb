```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from io import StringIO
```

```python
CSV_PATH = "../evaluation/evaluation_results_laberta_vulg_top1000.csv"

df = pd.read_csv(CSV_PATH)
pct_cols = ["Overall_Recall","Explicit_Recall","Implicit_Recall","Overall_MRR","Explicit_MRR","Implicit_MRR"]
df[pct_cols] = df[pct_cols] * 100

sns.set_theme(style="whitegrid", font="serif")
PALETTE = {"Overall": "#0B0B0B", "Explicit": "#C11A9A", "Implicit": "#F27209"}
MARKERS = {"Overall": "o", "Explicit": "s", "Implicit": "^"}
LINE_KW = dict(linewidth=2.2, markersize=8, markeredgewidth=1.4, markeredgecolor="white")

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))

# Recall@k
ax = axes[0]
for label, col in [("Overall","Overall_Recall"),("Explicit","Explicit_Recall"),("Implicit","Implicit_Recall")]:
    ax.plot(df["k"], df[col], label=label, color=PALETTE[label], marker=MARKERS[label], **LINE_KW)
    ax.annotate(f"{df[col].iloc[-1]:.0f}%", xy=(df["k"].iloc[-1], df[col].iloc[-1]),
                xytext=(12,0), textcoords="offset points", va="center",
                fontsize=9, color=PALETTE[label], fontweight="bold")

ax.axhline(y=40, color="#2C5F8A", linestyle=":", linewidth=1.4, alpha=0.55)
ax.set_title("Recall@k", fontsize=12, fontweight="semibold", pad=10)
ax.set_xlabel("k (log scale)", fontsize=10)
ax.set_ylabel("Recall (%)", fontsize=10)
ax.set_xscale("log")
ax.set_xticks(df["k"])
ax.get_xaxis().set_major_formatter(mticker.ScalarFormatter())
ax.set_ylim(0, 100)
ax.yaxis.set_major_formatter(mticker.PercentFormatter(decimals=0))
ax.grid(axis="y", alpha=0.35, linestyle="--")
ax.grid(axis="x", alpha=0.2, linestyle=":")

# MRR@k
ax2 = axes[1]
for label, col in [("Overall","Overall_MRR"),("Explicit","Explicit_MRR"),("Implicit","Implicit_MRR")]:
    ax2.plot(df["k"], df[col], label=label, color=PALETTE[label], marker=MARKERS[label], **LINE_KW)
    ax2.annotate(f"{df[col].iloc[-1]:.1f}%", xy=(df["k"].iloc[-1], df[col].iloc[-1]),
                 xytext=(12,0), textcoords="offset points", va="center",
                 fontsize=9, color=PALETTE[label], fontweight="bold")

ax2.set_title("MRR@k", fontsize=12, fontweight="semibold", pad=10)
ax2.set_xlabel("k (log scale)", fontsize=10)
ax2.set_ylabel("MRR (%)", fontsize=10)
ax2.set_xscale("log")
ax2.set_xticks(df["k"])
ax2.get_xaxis().set_major_formatter(mticker.ScalarFormatter())
ax2.yaxis.set_major_formatter(mticker.PercentFormatter(decimals=1))
ax2.grid(axis="y", alpha=0.35, linestyle="--")
ax2.grid(axis="x", alpha=0.2, linestyle=":")

# Create one shared legend under the suptitle
handles, labels = axes[0].get_legend_handles_labels()

fig.legend(
    handles, labels,
    loc="upper center",
    bbox_to_anchor=(0.5, 0.93),  
    ncol=3,
    frameon=False,
    fontsize=10
)

plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.savefig("../figures/recall_at_k_laberta.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch

models = {
    "LaBERTa\n(bi-encoder)": {
        "path": "../retrieval/laberta-vulg/retrieval_results_laberta_vulg.pkl",
        "type": "biencoder",
        "color": "#2C5F8A",
    },
    "SPhilBERTa\n(bi-encoder)": {
        "path": "../retrieval/sphilberta/retrieval_results_sphilberta.pkl",
        "type": "biencoder",
        "color": "#C1531A",
    },
        "LaBERTa + PhilBERTa\n(bi-enc + cross-enc)": {
        "path": "../retrieval/laberta-vulg-philberta/retrieval_results_laberta_vulg_philberta_reranked.pkl",
        "type": "probability",         
        "color": "#8B4A9B",
    },
    "LociSimiles\n(bi-enc + cross-enc)": {
        "path": "../retrieval/locisimiles/retrieval_results_locisimiles_judgement_k100.pkl",
        "type": "probability",  
        "color": "#4A9B6F",
    },
}

# Load scores to probabilities or cosine 
for name, cfg in models.items():
    with open(cfg["path"], "rb") as f:
        data = pickle.load(f)

    scores = [s for cands in data.values() for _, s in cands]

    if cfg["type"] == "probability":
        cfg["score_label"] = "Classifier probability P(positive)"
    else:
        cfg["score_label"] = "Cosine similarity"

    cfg["scores"] = scores
    cfg["data"]   = data

sns.set_theme(style="whitegrid", font="serif")
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, (name, cfg) in enumerate(models.items()):
    scores     = cfg["scores"]
    color      = cfg["color"]
    data       = cfg["data"]
    ax = axes[i]
    ax.hist(scores, bins=100, color=color, alpha=0.8, edgecolor="white", linewidth=0.3)
    ax.axvline(np.median(scores), color="black", linestyle="--",
               linewidth=1.2, alpha=0.6, label=f"Median={np.median(scores):.2f}")
    ax.set_xlabel(cfg["score_label"], fontsize=9)
    ax.set_ylabel("Count", fontsize=9)
    ax.set_title(f"{name} — score distribution", fontsize=10, fontweight="semibold")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig("../figures/score_distributions_all_models.png", bbox_inches="tight", dpi=300)
plt.show()

# ── Summary table ──────────────────────────────────────────────────────────
print(f"{'Model':<45} {'Min':>7} {'Median':>7} {'Max':>7}  Type")
print("-" * 75)
for name, cfg in models.items():
    s = cfg["scores"]
    print(f"{name.replace(chr(10), ' '):<45} {min(s):>7.3f} {np.median(s):>7.3f} {max(s):>7.3f}  {cfg['type']}")
```
